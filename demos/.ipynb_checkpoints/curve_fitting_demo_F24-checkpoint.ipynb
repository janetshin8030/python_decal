```python
import matplotlib.pyplot as plt
import scipy.optimize as fit
import numpy as np
```

```python
m = 0.5
b = 7
sigma = 1
n = 20
xdata = np.linspace(0, 10, 20)
ydata = (m * xdata + b) + np.random.standard_normal(size=n) * sigma
```

```python
plt.figure()
plt.scatter(xdata, ydata)
plt.show()
```

```python
def f(x, m, b):
    return m * x + b

p0 = [1, 7]

parameters, covariance_matrix = fit.curve_fit(f, xdata, ydata, p0) 
```

```python
parameters
```

```python
covariance_matrix
```

```python
m_error = covariance_matrix[0,0]
b_error = covariance_matrix[1,1]
```

```python
m_pred = parameters[0]
b_pred = parameters[1]
```

```python
print(m_error, m_pred)
```

```python
xpred = np.linspace(0, 10, 150)
y = f(xpred, m_pred, b_pred)

plt.figure()
plt.plot(xpred, y)
plt.scatter(xdata, ydata)
plt.show()
```
