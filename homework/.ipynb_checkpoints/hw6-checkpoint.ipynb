```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def generate_sinusoid(x, A, w):
    return A * np.sin(w * x)
```

```python
x=np.linspace(0,2*np.pi, 1000)
A_value= np.array([1,2,3,4,5])
w_value= np.array([1,2,3,4,5])
```

```python
for i in range(len(A_value)):
    A= A_value[i]
    w= w_value[i]
    y = A * np.sin(w * x)
    plt.plot(x, y, label=f'A={A}, w={w}')

plt.legend(loc="upper right")
```
