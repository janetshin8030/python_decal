```python
import numpy as np
import matplotlib as plt
import pandas as pd
```

```python
df = pd.read_csv("global_air_quality.csv")
```

```python
df["PM25_Value"] = df["FactValueNumeric"]
print(df[["PM25 Value" , "FactValueNumeric"]])
```

```python
df["PM25_Value"] = df["PM25_Value"].fillna(0)
average_pm25 = df.groupby("ParentLocation")["PM25_Value"].mean()
print(average_pm25)
```

```python
max_pm25_location = average_pm25.idxmax()
print(max_pm25_location)
```

```python
df["Average PM25"] = df["PM25 Value"].mean() 
print(df.columns)
```

```python
df_processed = df[[ "Location type","ParentLocation", "PM25 Value", "Average PM25"]]
df_processed.to_csv('processed_air_quality.csv')
```
